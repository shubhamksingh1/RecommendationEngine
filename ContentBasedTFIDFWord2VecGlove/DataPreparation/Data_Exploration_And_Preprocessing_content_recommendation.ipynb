A comparison of the most common words and the default English stop words will give us a list of words that need to be added to a custom stop word list.

```
def custom_stop_list(test_str):
    #Identify common words
    exception_word = ['ai','it','hr','cx','hp']
    freq_i = pd.Series(''.join(test_str).split()).value_counts()
    most_common_list = list(freq_i[:100].keys())
    least_common_list = list(freq_i[-100:].keys())
    custom_stop_list = list(set(most_common_list+least_common_list))
    mod_custom_stop_list = [x for x in custom_stop_list if len(x)<3 and x not in exception_word]
    return(mod_custom_stop_list)
    


```

```
print(custom_stop_list(df['asset_description'][0]))
```

```
df['custom_stop_desc'] = df['asset_description'].apply(custom_stop_list)
```

```
df['custom_stop_title'] = df['asset_title'].apply(custom_stop_list)
```

```
import nltk
nltk.download('stopwords')
```

```
from nltk.corpus import stopwords
nltk_stopwords = list(set(stopwords.words('english')))
```

```
print(nltk_stopwords)
```

```
list(set(nltk_stopwords + df['custom_stop_desc'][0]))
    
```

```
import re
```

```
nltk.download('wordnet') 
from nltk.stem.wordnet import WordNetLemmatizer
```

```
def stopwords_lemmatize_desc(i):
    #Remove punctuations
    text = re.sub('[^a-zA-Z]', ' ', df['asset_description'][i])
    stop_words = list(set(nltk_stopwords + df['custom_stop_desc'][i]))
    exception_word = ['ai','it','hr','cx','hp']
    stopwords = [w for w in stop_words if w not in exception_word]
    
    #Convert to lowercase
    text = text.lower()
    
    #remove tags
    text=re.sub("&lt;/?.*?&gt;"," &lt;&gt; ",text)
    
    # remove special characters and digits
    text=re.sub("(\\d|\\W)+"," ",text)
    
    ##Convert to list from string
    text = text.split()
    text = [ w for w in text if len(w)>1]

    #Lemmatisation
    lem = WordNetLemmatizer()
    text = [lem.lemmatize(word) for word in text if not word in  
            stopwords] 
    text = " ".join(text)
    return text

```

```
lem2 = WordNetLemmatizer()
lem2.lemmatize("harder")
```

```
corpus_desc = []
```

```
for i in range(0, df.shape[0]):
    text = stopwords_lemmatize_desc(i)
    corpus_desc.append(text)
```

```
df['asset_description_cleaned'] = corpus_desc
```

```
def stopwords_lemmatize_title(i):
    #Remove punctuations
    text = re.sub('[^a-zA-Z]', ' ', df['asset_title'][i])
    stop_words = list(set(nltk_stopwords + df['custom_stop_title'][i]))
    exception_word = ['ai','it','hr','cx','hp']
    stopwords = [w for w in stop_words if w not in exception_word]
    
    #Convert to lowercase
    text = text.lower()
    
    #remove tags
    text=re.sub("&lt;/?.*?&gt;"," &lt;&gt; ",text)
    
    # remove special characters and digits
    text=re.sub("(\\d|\\W)+"," ",text)
    
    ##Convert to list from string
    text = text.split()
    text = [ w for w in text if len(w)>1]

    #Lemmatisation
    lem = WordNetLemmatizer()
    text = [lem.lemmatize(word) for word in text if not word in  
            stopwords] 
    text = " ".join(text)
    return text

```

```
corpus_title=[]
```

```
for i in range(0, df.shape[0]):
    text = stopwords_lemmatize_title(i)
    corpus_title.append(text)
```

```
df['asset_title_cleaned'] = corpus_title
```
